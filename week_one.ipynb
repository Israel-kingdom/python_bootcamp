# Control flow-Loops and Conditionals
iteration - One time through the loop


```python
data = ['a', 'b', 'c', 'd', 'e']

for i in range(10):
  print(i, end=" ")

print("")
for letter in data:
  print(letter, end=" ")

print("")
for i, letter in enumerate(data):
  print(i, letter, end=" ")
```

```python
data = ['a', 'b', 'c', 'd', 'e']

i = 0

while i < len(data):
  print(data[i], end=" ")
  i += 1

while True:
  x = input()
  if x == "stop":
    break
  elif x == "special":
    continue
  else:
    print("it wasn't stop or special")
  print("printing data...")

```

# Guessing game
Generate a random number -- random module

do while - Python doesn't have this built in.  
  do something once -- possibly repeat it.  
  Ask us for a number. If we get it wrong, ask again.  


logic for right / wrong:
  guessed == correct value
  

```python
from random import randint
low = 0
high = 100
correct = randint(low, high)
print(f'Guess a number between {low} and {high}.')

while True:
  guess = int(input())

  if guess < correct:
    print("Guess higher")
  elif guess > correct:
    print("Guess lower")
  else:
    print("You win")
    break
  # do something
  #check to see if we should do it again
```
